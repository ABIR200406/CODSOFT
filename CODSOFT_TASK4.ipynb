```python
#spam_detection
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Data Collectioning And Preprocessing

```python
#loading the data from csv file to a pandas dataframe
raw_mail_data = pd.read_csv('/content/spam.csv',encoding = 'ISO-8859-1')

```

```python
print(raw_mail_data)
```

```python
#replace the null value with a null strings
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing the first 5 rows of the dataframe
mail_data.head()
```

```python
#checking the no. of rows and columns in the dataframe
mail_data.shape
```

Label Encoding

```python
 #label spam mail as 0; ham mail is 1
 mail_data.loc[mail_data['v1'] == 'spam', 'v1',] = 0
 mail_data.loc[mail_data['v1'] == 'ham', 'v1',] = 1
```

```python
#seperating the data as texts and label
X = mail_data['v2']
Y = mail_data['v1']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & test data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction

```python
#transform the text data to feature vectors that can be used as input to the logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

```

```python
X_train_feature = feature_extraction.fit_transform (X_train)
X_test_feature = feature_extraction.transform(X_test)

#convert y test and y train values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_feature)
```

Training the Model

Logistic Regression

```python
model= LogisticRegression()
```

```python
print(Y_train.dtype)
# Check for missing or null values
print(Y_train.isnull().sum())

# If necessary, convert Y_train to a numeric type (e.g., int) and handle missing values
Y_train = Y_train.astype(int)  # Assuming you want integer labels
Y_train = Y_train.fillna(0)

# training the logistic regression model with the training data
model.fit(X_train_feature, Y_train)
```

Evalutaing the trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_feature)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# Check the data type of Y_test
print(Y_test.dtype)

# Check for unique values in Y_test
print(Y_test.unique())

# If necessary, convert Y_test to a numeric type (e.g., int) and handle missing values, similar to what you did for Y_train
Y_test = Y_test.astype(int)  # Assuming you want integer labels
Y_test = Y_test.fillna(0)
```

```python
# prediction on test data

prediction_on_test_data = model.predict(X_test_feature)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a  blessing at all times"]

# convert text to feature vectors
input_data_feature = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_feature)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
